```python
# Todas las librerias que seran utilizadas
import matplotlib.pyplot as plt
from numpy import *
from matplotlib import rc
rc('text', usetex=True)
```

```python
# Se definen las constantes que se usan para la ecuación de Plank
k = 1.380649 * 10**(-23)
c = 299792458 
h = 6.62607015 * 10**(-34)
a = h*c/k
```

```python
# Se define la ecuación de Plank, para calcular el poder emisivo espectral con las variables de longitud de onda y la temperatura
def E(y,t):
    E = (8*pi*a*k/(y**5))/(e**(a/(y*t))-1)
    return(E) # Devuelve el valor del poder emisivo espectral
```

```python
T1 = 3500 # Se proporciona una temperatura (la minima)
t1 = '$%g ^{\circ}K$' %T1 # Se escribe esta temperatura en forma  latex para grados kelvin
T2 = 5758 # Se proporciona una temperatura (la maxima)
t2 = '$%g ^{\circ}K$' %T2 # Se escribe esta temperatura en forma  latex para grados kelvin
T3 = 5260 # Se proporciona una temperatura
t3 = '$%g ^{\circ}K$' %T3 # Se escribe esta temperatura en forma  latex para grados kelvin
T4 = 4000 # Se proporciona una temperatura
t4 = '$%g ^{\circ}K$' %T4 # Se escribe esta temperatura en forma  latex para grados kelvin
T5 = 4900 # Se proporciona una temperatura
t5 = '$%g ^{\circ}K$' %T5 # Se escribe esta temperatura en forma  latex para grados kelvin
m1 = 0.002898/T1 # Con la temperatura minima se calcula su longitud de onda que da el mayor poder de emision espectral
m2 = 0.002898/T2 # Con la temperatura maxima se calcula su longitud de onda que da el mayor poder de emision espectral
MF = arange(m2, m1, 10**(-9)) # Se da un rango un rango de longitudes de onda para calcular unicamente para las que dan el mayor poder de emision espectral 
F = arange(6*10**(-9), 2*10**(-6), 10**(-9)) # Se da un rango un rango de longitudes de onda para calcular la curva de poder de emision espectral con las temperaturas ya dadas (cubre cada nanometro hasta 2 micrometros)
MI = [] # Se haran las listas donde se colocara cada valor de poder de emision espectral
I1 = []
I2 = []
I3 = []
I4 = []
I5 = []
for i in MF: # Se calcula el poder de emision espectral por cada valor de longitud de onda "maxima" y la temperatura correspondiente a su maximo
    MI.append(E(i,0.002898/i)/1000) # Se divide entre 1000 para converit W/m^3 a kW/m^3
for i in F: # Se calcula los poderes de emision espectral por cada valor de longitud de onda y las temperaturas correspondientes
    I1.append(E(i,T1)/1000) # Se divide entre 1000 para converit W/m^3 a kW/m^3
    I2.append(E(i,T2)/1000)
    I3.append(E(i,T3)/1000)
    I4.append(E(i,T4)/1000)
    I5.append(E(i,T5)/1000)
MF = arange(m2*1000000, m1*1000000, 10**(-3)) # Se hace una nueva lista y se multiplica por 1000000 para transformar metros a micrometros
F = arange(6*10**(-3), 2, 10**(-3)) # Se hace una nueva lista para transformar metros a micrometros
plt.figure(figsize=(14,7)) # La figura tendra un tamaño de 14x7 pulgadas
plt.plot(F, I1, label = t1) # Se grafica la curva de temperatura 1 y se muestra su temperatura en una leyenda en forma latex
plt.plot(F, I2, label = t2) # Se grafica la curva de temperatura 2 y se muestra su temperatura en una leyenda en forma latex
plt.plot(F, I3, label = t3) # Se grafica la curva de temperatura 3 y se muestra su temperatura en una leyenda en forma latex
plt.plot(F, I4, label = t4) # Se grafica la curva de temperatura 4 y se muestra su temperatura en una leyenda en forma latex
plt.plot(F, I5, label = t5) # Se grafica la curva de temperatura 5 y se muestra su temperatura en una leyenda en forma latex
plt.plot(MF, MI, label = "Emisiones Máximas") # Se grafica la curva las longitudes de onda que dan poder emisivo espectral maximo con los rangos de temperatura de la minima a la maxima que se graficaron
plt.legend(loc=1) # Se colocan las leyendas en la posición superior derecha
plt.annotate('Máximos', (m2*1000000, E(m2,T2)/1000), xytext=(1,500), arrowprops=dict(facecolor='black', width=1, headwidth=5, shrink=0.05), horizontalalignment='left') # Se apunda en maximo de la temperatura mayor
plt.annotate('Máximos', (m1*1000000, E(m2,T1)/1000), xytext=(1,500), arrowprops=dict(facecolor='black', width=1, headwidth=5, shrink=0.05), horizontalalignment='left') # Se apunda en maximo de la temperatura menor
plt.title("Intencidad de la Radiacion de los  \"Cuerpos Negros\" ", fontsize=15) # Se da el titulo de la grafica
plt.xlabel("Longitud de Onda $\lambda \, (\mu m)$",fontsize=12) # Se pone en eje x el elemento que muestra y su unidad con latex
plt.ylabel("Poder Emisivo Espectral $(kW/m^3)$",fontsize=12) # Se pone en eje y el elemento que muestra y su unidad con latex
```
